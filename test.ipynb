```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# helper function: read image as RGB
def imread(img_path):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img

# helper function: display RGB image
def imshow(img, figsize=(10, 10)):
    plt.figure(figsize=figsize)
    plt.imshow(img)
    plt.show()
```

```python
# display an example image
# source: https://regalgentleman.com/blogs/blog/how-to-find-your-face-shape-for-men
img = imread('faces.png')
imshow(img)
```

```python
# create a face detector object
detector = FaceDetector('face_detector/')

# detect faces from img
faces = detector.detect(img)

# copy image to draw boxes around faces
img_copy = img.copy()

# loop through faces
for face in faces:
    
    # print detection confidence
    confidence = face['confidence']
    print(confidence)
    
    # display face sub-image
    imshow(face['img'], figsize=(3, 3))
    
    # draw a box around face
    (x, y, w, h) = face['box']
    cv2.rectangle(img_copy, (x, y), (x+w, y+h), (255, 0, 0), 2)
    # write confidence above each face
    cv2.putText(
        img_copy, 
        '{:.2f}%'.format(confidence*100), 
        (x, y-5), 
        cv2.FONT_HERSHEY_PLAIN, 
        1.5, 
        (255, 0, 0), 
        2,
    )

# display image with boxes around faces    
imshow(img_copy)
```
